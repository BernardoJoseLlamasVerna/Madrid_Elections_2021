Queremos traer los datos de las elecciones autonómicas de Madrid 2019 haciendo Scrapping de la siguiente página: https://resultados.elpais.com/elecciones/2019/autonomicas/12/

Instalamos beautifulsoup4: pip install beautifulsoup4
Intalamos lxml: pip install lxml
Instalamos requests library: pip install requests

```python
from bs4 import BeautifulSoup
import requests
```

```python
html_text = requests.get('https://resultados.elpais.com/elecciones/2019/autonomicas/12/').text
print(html_text)
```

```python
soup = BeautifulSoup(html_text, 'lxml')
```

```python
soup
```

```python
# municipios = soup.find_all('ul', class_='estirar')
# municipios
```

```python
# evita pillar los que tienen class estirar con algo más
municipios_list = soup.select("ul[class='estirar']")
municipios_list
```

```python
# listar todos los li y sacar su info:
for li in municipios_list:
    print(li.text, end="\n ")
```

```python
ul = soup.select('ul.estirar')[1]
ul
```

```python
links = [a['href'] for a in ul.select('a[href]')]
links
```

```python
lis = ul.find_all('li') 
lis
```

```python
urls=[]
texts=[]
for li in lis:
    for link in li.find_all('a'):
        urls.append(link.get('href'))
        texts.append(link.text)
```

```python
url = 'https://resultados.elpais.com/elecciones/2019/autonomicas/12/'
results = []
for li in lis:
    for link in li.find_all('a'):
        local_result = {}
        local_result['municipio'] = link.text
        local_result['link'] = url+link.get('href')
        results.append(local_result)
        
results
```
